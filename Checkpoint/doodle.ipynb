```python
from configs import *
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import numpy as np
```

Modified FitzHugh-Nagumo model:

```python
k = 0.001
b = 0.2
l = 1
def mfn(t, z):
    V, w = z
    return [
        -k * (V - V_REST) * (V - V_TH) * (V - V_PEAK) - w,
        V - V_REST - b * w,
    ]


T = 30
V_0 = -30
w_0 = 100

sol = solve_ivp(mfn, [0, T], [V_0, w_0], dense_output=True)

t = np.linspace(0, T, 500)
plt.plot(t, sol.sol(t).T)
plt.xlabel("t")
plt.legend(["V", "w"], shadow=True)
plt.title("Modified FitzHugh-Nagumo model")
plt.show()
```

```python
plt.plot(sol.sol(t)[0], sol.sol(t)[1])
plt.plot(V_REST, 0, "o", color="green")
plt.plot(V_0, w_0, "o", color="red")
plt.xlabel("V")
plt.ylabel("w")
plt.title("Phase space")
plt.show()
print("Finalni V i w su:", sol.sol(t)[0][-1], sol.sol(t)[1][-1])
```

FitzHugh-Nagumo model:

```python
a = 0.7
b = 0.8
tau = 12.5
R = 0.0
I_ext = 0.5


def fn(t, z):
    V, w = z
    return [
        V - V**3 / 3 - w + R * I_ext,
        1 / tau * (V + a - b * w),
    ]


T = 50
sol = solve_ivp(fn, [0, T], [-1.2, 0], dense_output=True)
t = np.linspace(0, T, 500)
plt.plot(t, sol.sol(t).T)
plt.xlabel("t")
plt.legend(["V", "w"], shadow=True)
plt.title("FitzHugh-Nagumo model")
plt.show()
sol.sol(t)[0][-1]
```

My model:

```python
def model(t, z):
    V, w = z
    return [
        -k * (V - V_REST) * (V - V_TH) * (V - V_PEAK)
        + k * (V - V_REST) * (V - V_PEAK) * w,
        l * (V - V_REST - b * w),
    ]


T = 20
V_0 = 10
w_0 = 0
sol = solve_ivp(model, [0, T], [V_0, w_0], dense_output=True)
t = np.linspace(0, T, 500)
plt.plot(t, sol.sol(t).T)
plt.xlabel("t")
plt.legend(["V", "w"], shadow=True)
plt.title("Modified FitzHugh-Nagumo model")
plt.show()
sol.sol(t)[0][-1]
```
