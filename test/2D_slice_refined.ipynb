# Testing the model on refined 2D slice of a heart

```python
domain = import_mesh('data/refined_slice.xdmf')
x = ufl.SpatialCoordinate(domain)
cell_model = ReparametrizedFitzHughNagumo()


class Model(BidomainModel):
    def __init__(
        self,
    ):
        pass

    def initial_V_m(self):
        return lambda x: (x[0] + 1.5) ** 2 + (x[1] - 1.5) ** 2 < 0.2**2, 0.0

    def ischemia(self):
        return (
            lambda x: (x[0] - 1.3) ** 2 / 2 + (x[1] - 2.2) ** 2 / 15 < 0.2**2,
            self.M_i/10,
            self.M_e/10,
        )
    
model = Model()
model.setup(domain, [0,0,0], [0,0,0])
```

```python
model.solve(
    T=1000,
    steps=500,
    cell_model=cell_model,
    gif_name="2D_slice_refined.gif",
    signal_point=[-1.5, -0.7, 4.25],
)
```
