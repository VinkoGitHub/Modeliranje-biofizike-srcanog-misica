```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 1, 200)
V_m = np.linspace(-50, 100, 200)

#  ovo mozda ubaciti u config file
V_min = -80 # mV
V_max = 20 # mV
tau_in = 0.01
tau_out = 0.1

tau_open = 1
tau_close = 1
V_gate = 3
```

```python
def I_ion(V_m, w):
    I_in = -w/tau_in*(V_m - V_min)**2*(V_max - V_m)/(V_max - V_min)
    I_out = 1/tau_out*(V_m - V_min)/(V_max - V_min)
    return I_in + I_out

def g(V_m, w, V_gate = V_gate):
    if(V_m < V_gate):
        return w/tau_open - 1/tau_open/(V_max - V_min)**2
    elif(V_m >= V_gate):
        return w/tau_close
```

```python
plt.figure(figsize=(5,3))
plt.plot(V_m, I_ion(V_m, 50), color = 'firebrick', label = '$I_{ion}(V_m, w)$')
plt.plot(V_m, [g(v,50) for v in V_m], color = 'darkblue', label = '$g(V_m, w)$')
plt.legend()
```
